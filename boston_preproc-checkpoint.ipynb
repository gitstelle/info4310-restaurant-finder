```python
filter1 = []
for x in data['search category']:
    if x in ['coffee','bakeries','cafes', 'donuts']:
        filter1.append('cb')
    elif x == 'ethnicmarkets':
            filter1.append('market')
    else:
        filter1.append('restaurant')
```

```python
data['filter1'] = filter1
```

In the original dataset, "categories" column is a string of a double bracketed list. The following lambda function takes the first object of that list into "filter2" column.

```python
#df = df.apply(lambda x: x.str.replace(',', ''))
data['filter2'] = data['categories'].apply(lambda x: x.replace('[','').replace(']','').replace('"','').split(',')[0].strip())
```

```python
filter2unique = data.filter2.unique()
filter2unique
```

```python
newfilter1 = []

# filter criteria
cb = ['Breakfast & Brunch','Coffee & Tea','Bakeries','Cafes','Donuts'] # cafes, bakeries
market = ['Delis','Grocery'] # markets
dd = ['Juice Bars & Smoothies','Bubble Tea','Ice Cream & Frozen Yogurt','Desserts','Bars','Lounges','Cocktail Bars'] #drinks, desserts

for x in range(len(data)):
    row = data.iloc[x,:]
    if (row['filter2'] in cb) or (row['filter1']=='cb'):
        newfilter1.append('cb')
    elif (row['filter2'] in market) or (row['filter1'] == 'market'):
        newfilter1.append('market')
    elif row['filter2'] in dd:
        newfilter1.append('dd')
    else:
        newfilter1.append('restaurant')
```

```python
data['filter1'] = newfilter1
```

```python
filter2new = []
am = ['American (Traditional)','American (New)']
jpn = ['Sushi Bars','Ramen']
itl = ['Pizza']
cn = ['Dim Sum','Cantonese']

for x in range(len(data)):
    row = data.iloc[x,:]
    f2 = row['filter2']
    if f2 in am:
        filter2new.append('American')
    elif f2 in jpn:
        filter2new.append('Japanese')
    elif f2 in itl:
        filter2new.append('Italian')
    elif f2 in cn:
        filter2new.append('Chinese')
    else:
        filter2new.append(f2)
```
